# О данных

Признаки:


*   job_id - нумерация, id каждого предложения
*   title - название объявления
*   location - географическое расположение работы из объявления
*   department - корпоративный отдел
*   salary_range - ориентировачный диапазон зарплаты
*   company_profile - краткое описание компании
*    description - описание работы из объявления
* requirements - необходимые требования для работы
* benefits - предлагаемые работадателем льготы
* **telecommuting** - 1 (true) для предложений дистанционной работы, 0, соответственно, для не дистанционной
* **has_company_logo** - есть ли у компании лого (1 если да)
* **has_questions** - 1, если есть отборочные вопросы
* employment_type - полный рабочий день (full-type), частичная занятость (part-time), контракт (contract)
* required_experience - какой опыт требуется (executive, entry level, intern)
* required_education - какой уровень обарзования требуется
* industry - какая сфера работы (IT, здоровье и т.д.)
* function - задача (consulting, engineering)
* **fraudulent** - является ли фейковым/скам объявление (1 если да, 0 если нет)
